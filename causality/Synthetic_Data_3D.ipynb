```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import sparse
from scipy.sparse.linalg import spsolve
from mpl_toolkits.mplot3d import Axes3D

# Parameters
Lx = 10.0   # Length of the domain in the x direction
Ly = 10.0   # Length of the domain in the y direction
T = 200.0     # Total simulation time
Nx = 50     # Number of spatial points in the x direction
Ny = 50     # Number of spatial points in the y direction
Nt = 1000    # Number of time steps
alpha = 0.01  # Thermal diffusivity
beta = 0.01   # Diffusion coefficient for wind equation
gamma = 0.1   # Damping coefficient for v
delta = 0.5   # Amplitude of external forcing term in v
omega = 2.0   # Frequency of external forcing term in v
eta = 0.3     # Coefficient influencing w in the wind equation
zeta = 3.5    # Coefficient influencing v in the new equation for w
theta = 0.3   # Coefficient influencing u in the new equation for w

# Spatial and temporal discretization
x = np.linspace(0, Lx, Nx)
y = np.linspace(0, Ly, Ny)
t = np.linspace(0, T, Nt)
dx = x[1] - x[0]
dy = y[1] - y[0]
dt = t[1] - t[0]

# Initialize solution arrays
p = np.zeros((Nt, Nx, Ny))
u = np.zeros((Nt, Nx, Ny))
v = np.zeros((Nt, Nx, Ny))
w = np.zeros((Nt, Nx, Ny))
w_cf = np.zeros((Nt, Nx, Ny))

# Initial conditions for u, v, and w
X, Y = np.meshgrid(x, y)
p[0, :, :] = np.exp(-((X - Lx/2)**2 + (Y - Ly/2)**2) / 4)
u[0, :, :] = np.exp(-((2*X - Lx/2)**2 + (1.2*Y - Ly/2)**2) / 6)
v[0, :, :] = np.sin(np.pi * X / Lx) * np.sin(np.pi * Y / Ly)
w[0, :, :] = np.zeros((Nx, Ny))
w_cf[0, :, :] = np.zeros((Nx, Ny))

# Create sparse matrices for finite difference schemes
A_heat = sparse.eye(Nx * Ny) * (1 + 2 * alpha * (dx**2 + dy**2) / dt)
A_heat -= sparse.eye(Nx * Ny, k=1) * alpha * dx**2 / dt
A_heat -= sparse.eye(Nx * Ny, k=-1) * alpha * dx**2 / dt
A_heat -= sparse.eye(Nx * Ny, k=Nx) * alpha * dy**2 / dt
A_heat -= sparse.eye(Nx * Ny, k=-Nx) * alpha * dy**2 / dt

A_wind_v = sparse.eye(Nx * Ny) * (1 + 2 * beta * (dx**2 + dy**2) / dt)
A_wind_v -= sparse.eye(Nx * Ny, k=1) * beta * dx**2 / dt
A_wind_v -= sparse.eye(Nx * Ny, k=-1) * beta * dx**2 / dt
A_wind_v -= sparse.eye(Nx * Ny, k=Nx) * beta * dy**2 / dt
A_wind_v -= sparse.eye(Nx * Ny, k=-Nx) * beta * dy**2 / dt

A_wind_w = sparse.eye(Nx * Ny) * (1 + 2 * alpha * (dx**2 + dy**2) / dt)
A_wind_w -= sparse.eye(Nx * Ny, k=1) * alpha * dx**2 / dt
A_wind_w -= sparse.eye(Nx * Ny, k=-1) * alpha * dx**2 / dt
A_wind_w -= sparse.eye(Nx * Ny, k=Nx) * alpha * dy**2 / dt
A_wind_w -= sparse.eye(Nx * Ny, k=-Nx) * alpha * dy**2 / dt

# Time-stepping loop
for n in range(1, Nt):
    # Update u, v, and w using the modified system
    b_heat = p[n-1, :, :].flatten() * (1 - 2 * alpha * (dx**2 + dy**2) / dt)
    p[n, :, :] = spsolve(A_heat, b_heat).reshape((Nx, Ny))

    b_heat1 = u[n-1, :, :].flatten() * (1 - 2 * alpha * (dx**2 + dy**2) / dt)
    u[n, :, :] = spsolve(A_heat, b_heat).reshape((Nx, Ny))

    b_wind_v = v[n-1, :, :].flatten() * (1 - 2 * beta * (dx**2 + dy**2) / dt) - gamma * v[n-1, :, :].flatten() * dt + delta * (p[n, :, :].flatten()+u[n, :, :].flatten()) * np.sin(omega * t[n]) #- eta * w[n-1, :, :].flatten()
    v[n, :, :] = spsolve(A_wind_v, b_wind_v).reshape((Nx, Ny))

    b_wind_w = w[n-1, :, :].flatten() * (1 - 2 * alpha * (dx**2 + dy**2) / dt) + zeta * (v[n-1, :, :].flatten()**2) * dt + theta * p[n, :, :].flatten()
    w[n, :, :] = spsolve(A_wind_w, b_wind_w).reshape((Nx, Ny))

v_cf = np.copy(v)
v_cf[:, :, 10:20] = v_cf[:, :, 10:20] * eta

for n in range(1, Nt):
    b_wind_w_cf = w_cf[n-1, :, :].flatten() * (1 - 2 * alpha * (dx**2 + dy**2) / dt) + zeta * (v_cf[n-1, :, :].flatten()**2) * dt + theta * u[n, :, :].flatten()
    w_cf[n, :, :] = spsolve(A_wind_w, b_wind_w_cf).reshape((Nx, Ny))
```

```python
# Saving the Data
print(p.shape,u.shape,v.shape,w.shape,w_cf.shape)
p = p.reshape(1000, 50, 50, 1)
u = u.reshape(1000, 50, 50, 1)
v = v.reshape(1000, 50, 50, 1)
w = w.reshape(1000, 50, 50, 1)
w_cf = w_cf.reshape(1000, 50, 50, 1)
data = np.concatenate([p, u, v, w, w_cf], 3)
print(data.shape)
np.save("synthetic_spatiotemporal_data_without_spillover.npy", data)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.sparse import diags
from scipy.sparse.linalg import spsolve

# Parameters
Lx = 10.0   # Length of the domain in the x direction
Ly = 10.0   # Length of the domain in the y direction
T = 5.0     # Total simulation time
Nx = 50     # Number of spatial points in the x direction
Ny = 50     # Number of spatial points in the y direction
Nt = 200    # Number of time steps
v_x = 1.0   # X-component of the wind velocity field
v_y = 1.0   # Y-component of the wind velocity field

# Spatial and temporal discretization
x = np.linspace(0, Lx, Nx)
y = np.linspace(0, Ly, Ny)
t = np.linspace(0, T, Nt)
dx = x[1] - x[0]
dy = y[1] - y[0]
dt = t[1] - t[0]

# Initialize solution array for wind speed
v = np.zeros((Nt, Nx, Ny))

# Initial condition for wind speed
X, Y = np.meshgrid(x, y)
v[0, :, :] = np.sin(np.pi * X / Lx) * np.sin(np.pi * Y / Ly)

# Time-stepping loop
for n in range(1, Nt):
    # Construct sparse matrices for central difference schemes
    D2x = diags([-1, 2, -1], [-1, 0, 1], shape=(Nx, Nx), format='csr') / dx**2
    D2y = diags([-1, 2, -1], [-1, 0, 1], shape=(Ny, Ny), format='csr') / dy**2

    # Advection term for wind
    v_x_partial_x = (v[n-1, :, :] - np.roll(v[n-1, :, :], 1, axis=1)) / dx
    v_y_partial_y = (v[n-1, :, :] - np.roll(v[n-1, :, :], 1, axis=0)) / dy

    # Update v using the 2D advection equation for wind
    v_advection = -v_x * dt * v_x_partial_x - v_y * dt * v_y_partial_y
    v[n, :, :] = spsolve(diags([1], [0], shape=(Nx*Ny, Nx*Ny), format='csr') - dt * v_x * v_x_partial_x - dt * v_y * v_y_partial_y, v[n-1, :, :].flatten() + v_advection.flatten()).reshape((Nx, Ny))

# Plot the results for wind speed
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
X, Y = np.meshgrid(x, y)
ax.plot_surface(X, Y, v[0, :, :], cmap='viridis', rstride=5, cstride=5, alpha=0.7)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Wind Speed')
ax.set_title('2D Advection Equation (Wind Speed) at t=0')

plt.show()
```
